```python
def count_LOC(file):
    # Initialize a counter for the lines of code
    line_count = 0

    # Open the file in read mode
    with open(file, 'r') as file:
        # Iterate over each line in the file
        for line in file:
            # Increment the line count
            line_count += 1

    return line_count
```

```python
NE_cases = []
for pr_nr, project in pr_project_pair.items():
    
    NE ={}

    T_CPatches = 0
    T_GPatches = 0
    T_CLoc = 0
    T_GLoc = 0

    root_directory = f'../../../src/devchatgpt/notebooks/Repos_files/1/RepoData/{project}/'

    chat_files_directory = f'{root_directory}{pr_nr}/chatgpt/'
    patch_files_directory = f'{root_directory}{pr_nr}/github/'

    if not os.path.exists(chat_files_directory):
        for patch_file_name in [file for file in os.listdir(patch_files_directory) if not file.startswith('.')]:
            T_GPatches = T_GPatches + 1
            patch_file_path = os.path.join(patch_files_directory, patch_file_name)
            T_GLoc = T_GLoc + count_LOC(patch_file_path)

        NE = {
            "PR_No": pr_nr,
            "Project": f'{project}/{pr_nr}',
            "T_GPatches": T_GPatches,
            "T_GLOC": T_GLoc,
            "Ratio_GL_GP": math.ceil(T_GLoc/T_GPatches),
        }
        
        NE_cases.append(NE)
    # sys.exit()
        
```

```python
result = []
for pr_nr, project in pr_project_pair.items():
    
    data = {}

    T_CPatches = 0
    T_GPatches = 0
    T_CLoc = 0
    T_GLoc = 0

    root_directory = f'../../../src/devchatgpt/notebooks/Repos_files/1/RepoData/{project}/'

    chat_files_directory = f'{root_directory}{pr_nr}/chatgpt/'
    patch_files_directory = f'{root_directory}{pr_nr}/github/'

    if not os.path.exists(chat_files_directory):
        continue
    for chat_file_name in [file for file in os.listdir(chat_files_directory) if not file.startswith('.')]:
        T_CPatches = T_CPatches + 1
        chat_file_path = os.path.join(chat_files_directory, chat_file_name)
        T_CLoc = T_CLoc + count_LOC(chat_file_path)
        # T_CLoc = T_CLoc + 1
        # print(count_LOC(chat_file_path))
    for patch_file_name in [file for file in os.listdir(patch_files_directory) if not file.startswith('.')]:
        T_GPatches = T_GPatches + 1
        patch_file_path = os.path.join(patch_files_directory, patch_file_name)
        T_GLoc = T_GLoc + count_LOC(patch_file_path)
        # T_GLoc = T_GLoc + 1

    data = {
        "PR_No": pr_nr,
        "Project": f'{project}/{pr_nr}',
        "T_GPatches": T_GPatches,
        "T_GLOC": T_GLoc,
        "Ratio_GL_GP": math.ceil(T_GLoc/T_GPatches),
        "T_CPatches": T_CPatches,
        "T_CLOC": T_CLoc,
        "Ratio_CL_CP": math.ceil(T_CLoc/T_CPatches),
    }

    result.append(data)
    # sys.exit()
        
```
